```python
def r_squared(y_true, y_predict):
    y_true = tf.cast(y_true, tf.float32)
    y_predict = tf.cast(y_predict, tf.float32)
    SS_res  = K.sum(K.square(y_true - y_predict))
    SS_tot = K.sum(K.square(y_true- K.mean(y_true)))
    return 1- (SS_res/(SS_tot+ K.epsilon()))
```

```python
import tensorflow as tf
from tensorflow.keras.metrics import MeanSquaredError
from tensorflow.keras.losses import MeanAbsoluteError

# Define RMSE as a custom Keras metric
class RootMeanSquaredError(tf.keras.metrics.Metric):
    def __init__(self, name='rmse', **kwargs):
        super(RootMeanSquaredError, self).__init__(name=name, **kwargs)
        self.mse = MeanSquaredError()

    def update_state(self, y_true, y_pred, sample_weight=None):
        self.mse.update_state(y_true, y_pred, sample_weight)

    def result(self):
        return tf.sqrt(self.mse.result())

    def reset_states(self):
        self.mse.reset_states()
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, Model

# Custom attention layer
class AttentionLayer(layers.Layer):
    def __init__(self, **kwargs):
        super(AttentionLayer, self).__init__(**kwargs)

    def build(self, input_shape):
        self.W = self.add_weight(
            shape=(input_shape[-1], 1),
            initializer="glorot_uniform",
            trainable=True
        )
        super(AttentionLayer, self).build(input_shape)

    def call(self, inputs):
        # Attention scores and softmax normalization
        scores = tf.matmul(inputs, self.W)  # shape: [batch_size, feature_size, 1]
        scores = tf.nn.softmax(scores, axis=1)  # normalize scores to sum to 1
        context = tf.reduce_sum(inputs * scores, axis=1)  # weighted sum of inputs
        return context

# Create image input layer
image_input = keras.Input(shape=(512, 512, 3), name='image_input')
x = layers.Conv2D(16, (3, 3), activation='relu')(image_input)
x = layers.MaxPooling2D((2, 2))(x)
x = layers.Conv2D(32, (3, 3), activation='relu')(x)
x = layers.MaxPooling2D((2, 2))(x)
x = layers.Conv2D(64, (3, 3), activation='relu')(x)
x = layers.MaxPooling2D((2, 2))(x)
x = layers.Flatten()(x)

# Attention over flattened output
x = layers.Reshape((-1, 64))(x)  # Adding a dimension to apply attention
attention_image = AttentionLayer()(x)  # Attention requires at least 2D input

image_output = layers.Dense(64, activation='relu')(attention_image)


tabular_shape = (train_csv_np.shape[1],)

# Create tabular input layer
tabular_input = keras.Input(shape=tabular_shape, name='tabular_input')  # Adjust the shape
reshaped_input = layers.Reshape((1, tabular_shape[0]))(tabular_input)

# Recurrent Neural Network for tabular data processing (using LSTM)
y = layers.LSTM(32, return_sequences=True)(reshaped_input)

# Apply attention to LSTM output
attention_tabular = AttentionLayer()(y)

tabular_output = layers.Dense(64, activation='relu')(attention_tabular)

# Concatenate the outputs from both branches
combined = layers.concatenate([image_output, tabular_output])

# Final dense layers for prediction
z = layers.Dense(64, activation='relu')(combined)
z = layers.Dropout(0.5)(z)  # Dropout for regularization
output = layers.Dense(6, name='output')(z)  # 6 outputs for regression
```

```python
# Create model
model = Model(inputs=[image_input, tabular_input], outputs=output)
model.compile(optimizer=Adam(learning_rate=1e-5), loss=MeanAbsoluteError(), metrics=[r_squared, 
                                                                       RootMeanSquaredError()])

model.summary()
```

```python
early_stopper = EarlyStopping(
    monitor='val_r_squared',
    mode='max',
    min_delta=0.001,
    patience=10,
    verbose=1,
    restore_best_weights=True
)
```

```python
history = model.fit(train_data, epochs=50, batch_size= 100, validation_data= val_data, callbacks=[early_stopper])
```

```python
# Plotting loss curves
plt.figure(figsize=(8, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Loss Curve')
plt.legend()
plt.show()
```

```python
# Evaluate the model and get the R^2 score
results = model.evaluate(test_data)
r2 = results[1]
```
